```python
# Calculate the total budget

budget_total = school_data["budget"].sum()
budget_total
```

```python
# Calculate the average math score

average_math_score = student_data["math_score"].mean()
average_math_score
```

```python
# Calculate the average reading score

average_reading_score = student_data["reading_score"].mean()
average_reading_score
```

```python
# Calculate the overall passing rate (overall average score)
# i.e. (avg. math score + avg. reading score)/2
# Create a list of the avg. math score and avg. reading score

overall_average_score = (average_math_score + average_reading_score)/2
overall_average_score
```

```python
# Calculate the percentage of students with a passing math score (70 or greater)
# Condition if column Math Score > or = 70, then P

math_score = school_data_complete[school_data_complete.Math_Score >= 70]
math_score.count()
```

```python
# Calculate the percentage of students with a passing math score (70 or greater)
# Percentage of students with passing math score =
# (Passed / Total) * 100

student_data["#passing_math"] = student_data["math_score"] >= 70
percent_passing_math = ((student_data["#passing_math"]).mean())*100
percent_passing_math
```

```python
# Calculate the percentage of students with a passing reading score (70 or greater)
# Condition if column Reading Score > or = 70, then P

student_data["#passing_reading"] = student_data["reading_score"] >= 70
percent_passing_reading = ((student_data["#passing_reading"]).mean())*100
percent_passing_reading
```

```python
# Create a dataframe to hold the above results
# Create a list of the column names

summary_results = [{"Total Schools": school_total, 
                    "Total Students": student_total,
                    "Total Budget": budget_total, 
                    "Average Math Score": round(average_math_score,2), 
                    "Average Reading Score": round(average_reading_score,2), 
                    "Percent Passing Math": round(percent_passing_math,2),
                    "Percent Passing Reading": round(percent_passing_reading,2),
                    "Overall Passing Average": round(overall_average_score,2)}]
summary_table = pd.DataFrame(summary_results)
summary_table
```

```python
# Create an overview summary table
#For this part, school_data_complete

school_data_complete.head(10).style.format({"Budget": "${:20,.2f}"})
school_data_complete
```

```python
# Conditional statement to contain passing grades higher than or equal to 70

school_data_complete["passing_math"] = school_data_complete["Math_Score"] >= 70
school_data_complete["passing_reading"] = school_data_complete["Reading_Score"] >= 70
school_data_complete
```

```python
# Use groupby by school_name

school_summary = school_data_complete.groupby(["school_name"]).mean()
school_summary["Per Student Budget"] = school_summary["Budget"]/school_summary["Size"]
school_summary["Percent Passing Math"] = round(school_summary["passing_math"]*100,2)
school_summary["Percent Passing Reading"] = round(school_summary["passing_reading"]*100,2)
school_summary["Percent Overall Passing Rate"] = round(((school_summary["passing_math"] + school_summary["passing_reading"])/2)*100,2)
school_summary
```

```python
# Relisting the school data table just to have it closer for viewing column names

school_data
```

```python
#Merge with school_data to get the school type and budget data

school_data_summary = pd.merge(school_summary, school_data, how="left", on=["school_name", "school_name"])
school_data_summary
```

```python
# Delet the duplicate columns 'size' and 'budget'

school_data_summary = pd.merge(school_summary, school_data, how="left", on=["school_name", "school_name"])
del school_data_summary['size']
del school_data_summary['budget']
school_data_summary
```

```python
#Merge with student_data to get the grade into the table

school_student_data_summary = pd.merge(school_data_summary, student_data, how="left", on=["school_name", "school_name"])
school_student_data_summary
```

```python
# Create a dataframe to hold the above results

school_student_data_summary_dataframe = pd.DataFrame({"School Name":  school_student_data_summary["school_name"],
                                "School Type": school_student_data_summary["type"],
                               "Total Students":school_student_data_summary["Size"],
                               "Total School Budget": school_student_data_summary["Budget"],
                               "Per Student Budget":school_student_data_summary["Per Student Budget"], 
                               "Average Math Score":round(school_student_data_summary["Math_Score"],2),
                               "Average Reading Score":round(school_student_data_summary["Reading_Score"],2), 
                               "Percent Passing Math": school_student_data_summary["Percent Passing Math"],
                               "Percent Passing Reading": school_student_data_summary["Percent Passing Reading"],
                               "Percent Overall Passing Rate": school_student_data_summary["Percent Overall Passing Rate"]})
school_student_data_summary_dataframe
```
